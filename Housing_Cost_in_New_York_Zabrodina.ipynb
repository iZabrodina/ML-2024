# Обработка текстовых признаков

```python
data_raw['Job Description'].dtype
```

Приводим текст к нижнему регистру:

```python
def lower(text):
    return text.lower()

data_raw['Text'] = data_raw['Job Description'].apply(lambda x:lower(x))
```

Убираем url и html ссылки:

```python
def remove_urls(text):
  return re.sub(r'https?://\S+|www\.\S+', '', text)

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_urls(x))

def remove_html(text):
  return re.sub(r'<.*?>', '', text)

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_html(x))

def remove_mention(x):
    return re.sub(r'@\w+','',x)

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_mention(x))

def remove_hash(x):
    return re.sub(r'#\w+','',x)

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_hash(x))

def remove_amp(x):
    return re.sub(r'\&amp','',x)

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_amp(x))
```

Убираем stopwords и пунктуацию:

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
STOPWORDS = set(stopwords.words('english'))

def punct_remove(text):
    return re.sub(r'[^\w\s]', '', text)

data_raw['Text'] = data_raw['Text'].apply(lambda x:punct_remove(x))

def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word not in STOPWORDS])

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_stopwords(x))
```

Убираем лишние пробелы:

```python
def remove_space(text):
    return re.sub(r"\s+"," ",text).strip()

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_space(x))
```

Выполняем стемминг - находим основы слова для заданного исходного слова:

```python
from nltk.stem import PorterStemmer

def stemming(text):
    return ' '.join([PorterStemmer().stem(w) for w in text.split()])

data_raw['Text'] = data_raw['Text'].apply(lambda x:stemming(x))
```

# 6. Разбить данные на обучающую и тестовую выборки

```python
X = data_raw.drop(['Job #','Job Description', 'Text', 'Initial Cost'], axis=1)
X.reset_index(drop=True, inplace=True)
X_text = X_text.toarray()
df = pd.DataFrame(X_text)
X = pd.concat([X, df], axis=1, join="inner")
X.columns = X.columns.astype(str)
y = data_raw['Initial Cost']
```

Разобьем данные на обучающую и тестовую выборки следующим образом: 80% - обучающая выборка, 20% - тестовая.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state = 42)

N_train, _ = X_train.shape
N_test,  _ = X_test.shape

print('Размеры обучающей и тестовой выборок: ', N_train, ' и ', N_test)
```

# 7. Запустить классификатор ближайших соседей. Вычислить ошибки на тестовой и обучающей выборках

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

k_arr = [1, 2, 3, 5, 10]

knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': k_arr}, cv=4)
grid.fit(X_train, y_train)

err_train = np.mean(y_train != grid.predict(X_train))
err_test  = np.mean(y_test  != grid.predict(X_test))

best_n_neighbors = grid.best_estimator_.n_neighbors

print('Лучшее количество соседей: ', best_n_neighbors)
print('Ошибка на обучающей выборке: ', err_train)
print('Ошибка на тестовой выборке: ', err_test)

# Построим матрицу рассогласования
from sklearn.metrics import ConfusionMatrixDisplay
fig = plt.figure(figsize=(17, 17))
ax = fig.add_subplot(1,1,1)
ConfusionMatrixDisplay.from_estimator(grid, X_test, y_test, cmap=plt.cm.Blues, ax=ax)
pass
```

1. Многослойный перцептрон:

```python
from sklearn.neural_network import MLPClassifier

# Создание модели многослойного перцептрона
mlp = MLPClassifier(hidden_layer_sizes=(100, 50, 25), activation='tanh', solver='lbfgs', max_iter=10000, random_state=42)

# Обучение модели
mlp.fit(X_train, y_train)

# Предсказание и вычисление ошибок
Y_train_predict = mlp.predict(X_train)
Y_test_predict = mlp.predict(X_test)

err_train = np.mean(y_train != Y_train_predict)
err_test = np.mean(y_test != Y_test_predict)

print('Ошибка на обучающей выборке: ', err_train)
print('Ошибка на тестовой выборке: ', err_test)

fig_mlp = plt.figure(figsize=(17, 17))
ax = fig_mlp.add_subplot(1,1,1)
ConfusionMatrixDisplay.from_estimator(mlp, X_test, y_test, cmap=plt.cm.Blues, ax=ax)
pass
```

2. Случайный лес:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf_arr = [10, 50, 90, 130, 150]

# Создание модели случайного леса с фиксированными значениями n_estimators и max_depth
rf = RandomForestClassifier(n_estimators=100)

# Обучение модели
grid_rf = GridSearchCV(rf, param_grid = {'max_depth': rf_arr}, cv=4)
grid_rf.fit(X_train, y_train)

# Предсказание и вычисление ошибок
Y_train_predict = grid_rf.predict(X_train)
Y_test_predict = grid_rf.predict(X_test)

err_train = np.mean(y_train != Y_train_predict)
err_test = np.mean(y_test != Y_test_predict)

best_depth = grid_rf.best_estimator_.max_depth

print('Наилучшая глубина поиска: ', best_depth)
print('Ошибка на обучающей выборке: ', err_train)
print('Ошибка на тестовой выборке: ', err_test)

fig_rf = plt.figure(figsize=(17, 17))
ax = fig_rf.add_subplot(1,1,1)
ConfusionMatrixDisplay.from_estimator(grid_rf, X_test, y_test, cmap=plt.cm.Blues, ax=ax)
pass
```

3. Метод градиентного бустинга:

```python
from sklearn.ensemble import GradientBoostingClassifier

# Создание и обучение модели градиентного бустинга
gbm = GradientBoostingClassifier(n_estimators=5, max_depth=10, random_state=42)
gbm.fit(X_train, y_train)

# Предсказание на обучающей и тестовой выборках
Y_train_predict = gbm.predict(X_train)
Y_test_predict = gbm.predict(X_test)

# Расчет ошибок на обучающей и тестовой выборках
err_train = np.mean(y_train != Y_train_predict)
err_test = np.mean(y_test != Y_test_predict)

# Вывод ошибок
print('Ошибка на обучающей выборке: ', err_train)
print('Ошибка на тестовой выборке: ', err_test)

fig_gbm = plt.figure(figsize=(17, 17))
ax = fig_gbm.add_subplot(1,1,1)
ConfusionMatrixDisplay.from_estimator(gbm, X_test, y_test, cmap=plt.cm.Blues, ax=ax)
pass
```

# Выводы по всем рассмотренным методам:
1. Метод ближайших соседей (KNN):
- Показывает наименьшую ошибку среди рассматриваемых методов, что свидетельствует о его эффективности при обработке малых объемов данных.
- Может столкнуться с проблемами при увеличении объема данных, так как KNN основывается на поиске ближайшего соседа, что может стать неэффективным при большом количестве точек.

2. Многослойный перцептрон (MLP):
- Демонстрирует высокие ошибки на обоих наборах данных, что может говорить о недостаточной мощности модели для данной задачи или о проблемах с переносом знаний на новые данные.
- Может быть более мощным инструментом при наличии достаточного количества данных, однако на маленьких выборках может возникнуть переобучение.

3. Случайный лес (RF):
- Показывает умеренные ошибки на обучающей выборке и значительные ошибки на тестовой выборке. Это может свидетельствовать о хороших навыках обобщения, но трудностей с точной работой на новых данных.
- Хорошо подходит для задач, где важны хорошие навыки обобщения, но может испытывать сложности с точной обработкой новых данных.

4. Метод градиентного бустинга (GBM):
- Демонстрирует самые высокие ошибки среди методов, что указывает на проблемах с обобщением и потенциальной проблемой переобучения.
- Может показать улучшение при увеличении объема данных, но пока может страдать от проблем с переносом знаний на новую информацию.

Вывод: KNN остается лучшим выбором для малого объема данных благодаря своей способности находить ближайшие точки и избегать проблем с переобучением. 
MLP может оказаться более эффективным при наличии большого объема данных, но сейчас страдает от недостатка мощности и возможной переоценки.
RF хорошо подходит для задач, где важна хорошая генерализация, но может столкнуться с трудностями при обработке новых данных.
GBM обладает потенциалом для роста при увеличении объема данных, но требует внимания к вопросам переноса знаний и возможного переобучения.
