# Обработка текстовых признаков

```python
data_raw['Job Description'].dtype
```

Приводим текст к нижнему регистру:

```python
def lower(text):
    return text.lower()

data_raw['Text'] = data_raw['Job Description'].apply(lambda x:lower(x))
```

Убираем url и html ссылки:

```python
def remove_urls(text):
  return re.sub(r'https?://\S+|www\.\S+', '', text)

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_urls(x))

def remove_html(text):
  return re.sub(r'<.*?>', '', text)

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_html(x))

def remove_mention(x):
    return re.sub(r'@\w+','',x)

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_mention(x))

def remove_hash(x):
    return re.sub(r'#\w+','',x)

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_hash(x))

def remove_amp(x):
    return re.sub(r'\&amp','',x)

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_amp(x))
```

Убираем stopwords и пунктуацию:

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
STOPWORDS = set(stopwords.words('english'))

def punct_remove(text):
    return re.sub(r'[^\w\s]', '', text)

data_raw['Text'] = data_raw['Text'].apply(lambda x:punct_remove(x))

def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word not in STOPWORDS])

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_stopwords(x))
```

Убираем лишние пробелы:

```python
def remove_space(text):
    return re.sub(r"\s+"," ",text).strip()

data_raw['Text'] = data_raw['Text'].apply(lambda x:remove_space(x))
```

Выполняем стемминг - находим основы слова для заданного исходного слова:

```python
from nltk.stem import PorterStemmer

def stemming(text):
    return ' '.join([PorterStemmer().stem(w) for w in text.split()])

data_raw['Text'] = data_raw['Text'].apply(lambda x:stemming(x))
```

# 6. Разбить данные на обучающую и тестовую выборки

```python
# Построение гистограммы
plt.figure(figsize=(16, 10))
plt.hist(data_raw['Initial Cost'], bins=50)
# Вращение текста на оси Y
for tick in plt.gca().get_xticklabels():
    tick.set_rotation(+90)
    tick.set_ha('right')
plt.title('Распределение стоимости квартир')
plt.xlabel('Стоимость')
plt.ylabel('Количество квартир')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python

data_raw['Initial Cost'] = data_raw['Initial Cost'].astype(int)

print(data_raw['Initial Cost'].head())

unique_classes = data_raw['Initial Cost'].nunique()

# Добавляем шаг для разделения данных на классы по цене
price_bins = [5000000, 7000000, 10000000, 20000000]
price_labels = ["Дешевые", "Средние", "Дорогие"]

# Функция для назначения класса на основе ценового диапазона
def assign_class(price):
    for i, price_bin in enumerate(price_bins):
        if price < price_bin:
            return price_labels[i]
    return price_labels[0]  

# Применяем функцию к столбцу 'Initial Cost'
data_raw['Class'] = data_raw['Initial Cost'].apply(assign_class)

# Проверяем, что класс был успешно назначен
print(data_raw[['Initial Cost', 'Class']].head())

X = data_raw.drop(['Job #','Job Description', 'Text', 'Initial Cost', 'Class'], axis=1)
X.reset_index(drop=True, inplace=True)
X_text = X_text.toarray()
df = pd.DataFrame(X_text)
X = pd.concat([X, df], axis=1, join="inner")
X.columns = X.columns.astype(str)
y = data_raw['Class']  
```

Разобьем данные на обучающую и тестовую выборки следующим образом: 80% - обучающая выборка, 20% - тестовая.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.85, test_size = 0.15, random_state = 42)

N_train, _ = X_train.shape
N_test,  _ = X_test.shape

print('Размеры обучающей и тестовой выборок: ', N_train, ' и ', N_test)
```

# 7. Запустить классификатор ближайших соседей. Вычислить ошибки на тестовой и обучающей выборках

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': [1]}, cv=4)
grid.fit(X_train, y_train)

err_train = np.mean(y_train != grid.predict(X_train))
err_test  = np.mean(y_test  != grid.predict(X_test))

best_n_neighbors = grid.best_estimator_.n_neighbors

print('Лучшее количество соседей: ', best_n_neighbors)
print('Ошибка на обучающей выборке: ', err_train)
print('Ошибка на тестовой выборке: ', err_test)

# Построение матрицы несоответствия
from sklearn.metrics import ConfusionMatrixDisplay
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1,1,1)
ConfusionMatrixDisplay.from_estimator(grid, X_test, y_test, cmap=plt.cm.Blues, ax=ax)
pass
```
